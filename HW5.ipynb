### Problem 1 Solution

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/Matt/desktop/student_class.csv')
gMedian = df.groupby('Majors').median()
gMedian.reset_index(inplace=True)
final = gMedian['Final Average']
locations = np.arange(gMedian.shape[0])
labs = ['BNFO', 'CS', 'IS', 'IT', 'MATH']
width = 0.25
plt.bar(locations-width, gMedian['Final'], width = width, label='Final', color='red')
plt.bar(locations, gMedian['Midterm'], width = width, label='Midterm', color='green')
plt.bar(locations+width, gMedian['Final Average'], width = width, label='Final Average', color='blue')
plt.xticks(locations, labels=labs)
plt.ylabel('Median Score')
plt.xlabel('Majors')
plt.title('Median Midterm, Final, and Final Average Scores')
plt.legend()
plt.show()
```

```python
#The IT major is doing pretty well, because the median throughout the midterm, final, and final avergae 
#is more solid than other majors, meaning is not a gap between the scores.
#Items: Midterm, Final, Final Average, and major
#Attributes: Midterm, Final, and Final average scores; majors 
#Marks: Bars
#Channels: Vertical Bars
```

### Problem 2 Solution

```python
final = df['Final Average']
fMedian = df['Final Average'].median()
bins = [40, 50, 60, 70, 80, 90, 100]
plt.hist(final, bins=bins)
plt.ylabel('Frequency')
plt.xlabel('Grades')
plt.title('Number of Student in Each Grade Bin')
plt.show()
```

```python
#The histogram is missing some sort of seperator for each bins, 
#also showing the meadian between those values would of been something nice to have on the graph.
#Regarding the distribution it seems to be normal distributed it.
#Items: Final Avergae
#Attributes: Final Average Scores
#Marks: Bins
#Channels: Vertical Bins
```

```python
plt.hist(final, bins=bins, edgecolor='black')
plt.ylabel('Frequency')
plt.xlabel('Grades')
plt.title('Number of Student in Each Grade Bin')
plt.axvline(fMedian, color='red', linewidth=1, label='Median')
plt.legend()
plt.show()
```
